```python
import pandas as pd

# Cargar el dataset desde la ruta correcta
df = pd.read_csv('C:/Users/dulce/Downloads/ds_salaries.csv')

# Ver las primeras filas del dataset
df.head()
```

```python
# Verificar si hay duplicados
duplicados = df.duplicated().sum()
print(f"Valores duplicados: {duplicados}")

# Eliminar duplicados si es necesario
df = df.drop_duplicates()
```

```python
# Ver los tipos de datos
print(df.dtypes)

# Cambiar tipos de datos si es necesario (esto depende del análisis que realices)
# Ejemplo:
# df['salary'] = df['salary'].astype(float)
```

```python
# Convertir valores categóricos a formato consistente (por ejemplo, en minúsculas)
df['job_title'] = df['job_title'].str.lower()

# Verificar las categorías después del ajuste
print(df['job_title'].unique())
```

```python
# Verificar valores faltantes
print(df.isnull().sum())

# Rellenar valores faltantes si es necesario
df['salary'] = df['salary'].fillna(df['salary'].mean())  # Ejemplo de relleno
```

```python
# Detectar salarios extremadamente altos o bajos
anomalies = df[(df['salary'] > 1_000_000) | (df['salary'] < 10_000)]
print(f"Valores anómalos detectados: {len(anomalies)}")

# Puedes eliminar o investigar los valores anómalos según lo que encuentres
df = df[(df['salary'] <= 1_000_000) & (df['salary'] >= 10_000)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograma de los salarios
plt.figure(figsize=(10, 6))
sns.histplot(df['salary'], bins=20, kde=True)
plt.title('Distribución de los Salarios')
plt.xlabel('Salario')
plt.ylabel('Frecuencia')
plt.show()

# Gráfico de barras para los tipos de empleo
plt.figure(figsize=(10, 6))
sns.countplot(y='job_title', data=df, order=df['job_title'].value_counts().index)
plt.title('Frecuencia de Tipos de Trabajo')
plt.xlabel('Frecuencia')
plt.ylabel('Título del Trabajo')
plt.show()
```

```python
# Relación entre salario y años de experiencia
plt.figure(figsize=(10, 6))
sns.scatterplot(x='experience_level', y='salary', data=df)
plt.title('Relación entre Nivel de Experiencia y Salario')
plt.xlabel('Nivel de Experiencia')
plt.ylabel('Salario')
plt.show()

# Gráfico de caja para salarios por país
plt.figure(figsize=(10, 6))
sns.boxplot(x='company_location', y='salary', data=df)
plt.title('Distribución de Salarios por País')
plt.xlabel('País')
plt.ylabel('Salario')
plt.xticks(rotation=90)
plt.show()
```

```python
# Resumen estadístico del dataset
df.describe()
```

```python
# Analizar tendencias en los datos
# Por ejemplo, puedes identificar si hay alguna tendencia en salarios por años de experiencia
tendencia = df.groupby('experience_level')['salary'].mean()
print("Salario promedio por nivel de experiencia:")
print(tendencia)
```
